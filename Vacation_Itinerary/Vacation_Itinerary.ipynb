```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from Config3 import gkey
import pandas as pd
import gmaps
gmaps.configure(api_key=gkey)
```

```python
weatherpy_vacation_df = pd.read_csv('../Vacation Search/WeatherPy_vacation.csv')
weatherpy_vacation_df
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
fig = gmaps.figure()
marker_locations = weatherpy_vacation_df[['Lat', 'Lng']]
hotel_info = [info_box_template.format(**row) for i, row in weatherpy_vacation_df.iterrows()]
markers = gmaps.marker_layer(marker_locations, info_box_content=hotel_info)
fig.add_layer(markers)
fig
```

```python
Starting_pt_Esmeraldas_df = weatherpy_vacation_df.loc[weatherpy_vacation_df['City'] == 'Esmeraldas']
Starting_pt_Esmeraldas_df
Second_pt_Porciuncula_df = weatherpy_vacation_df.loc[weatherpy_vacation_df['City'] == 'Porciuncula']
Third_pt_SaoJoao_df = weatherpy_vacation_df.loc[weatherpy_vacation_df['City'] == 'Sao joao da barra']
Fourth_pt_Arraial_df = weatherpy_vacation_df.loc[weatherpy_vacation_df['City'] == 'Arraial do cabo']
Ending_Esmeraldas_df = weatherpy_vacation_df.loc[weatherpy_vacation_df['City'] == 'Esmeraldas']
Third_pt_SaoJoao_df
```

```python
start = Starting_pt_Esmeraldas_df['Lat'].to_numpy()[0], Starting_pt_Esmeraldas_df['Lng'].to_numpy()[0]
start
stop1 = Second_pt_Porciuncula_df['Lat'].to_numpy()[0], Second_pt_Porciuncula_df['Lng'].to_numpy()[0]
stop2 = Third_pt_SaoJoao_df['Lat'].to_numpy()[0], Third_pt_SaoJoao_df['Lng'].to_numpy()[0]
stop3 = Fourth_pt_Arraial_df['Lat'].to_numpy()[0], Fourth_pt_Arraial_df['Lng'].to_numpy()[0]
end = Ending_Esmeraldas_df['Lat'].to_numpy()[0], Ending_Esmeraldas_df['Lng'].to_numpy()[0]
end
```

```python

trip_start = start
trip_stop_1 = stop1
trip_stop_2 = stop2
trip_stop_3 = stop3
trip_end = end
fig = gmaps.figure()
trip_of_4_cities = gmaps.directions_layer(trip_start, trip_end, waypoints=[trip_stop_1,trip_stop_2,trip_stop_3], travel_mode='DRIVING')
fig.add_layer(trip_of_4_cities)
fig
```

```python
trip_hotel_df = pd.concat((Starting_pt_Esmeraldas_df, Second_pt_Porciuncula_df, Third_pt_SaoJoao_df, Fourth_pt_Arraial_df), ignore_index=True)
trip_hotel_df
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
fig = gmaps.figure()
marker_locations = trip_hotel_df[['Lat', 'Lng']]
hotel_info = [info_box_template.format(**row) for i, row in trip_hotel_df.iterrows()]
markers = gmaps.marker_layer(marker_locations, info_box_content=hotel_info)
fig.add_layer(markers)
fig
```
