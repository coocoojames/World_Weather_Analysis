```python
lats = np.random.uniform(-90,90, size=2000)
lats
lngs = np.random.uniform(-90,90, size=2000)
lngs
lats_lngs = zip(lats,lngs)
coordinates = list(lats_lngs)
coordinates
```

```python
from citipy import citipy as cpy
cities = []
for coordinate in coordinates:
    city = cpy.nearest_city(coordinate[0], coordinate[1]).city_name
    cities.append(city)
```

```python
import json
from Config import api_key
import pandas as pd
import requests as r
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID="
request_url = f'{base_url}{api_key}&q={city}'
weather = r.get(request_url).json()
print(json.dumps(weather, indent=4, sort_keys=True))
```

```python
import time
city_weather_ls = []
for i, city in enumerate(cities):
    if (i % 50 == 0 and i >= 50):
        time.sleep(60)
    request_url = f'{base_url}{api_key}&q={city}'
    weather = r.get(request_url).json()
    print(f'Getting weather data for {city}.')
    try:
        city_weather_ls.append({'City':city,
                               'Country':weather['sys']['country'],
                               'Lat':weather['coord']['lat'],
                               'Lng':weather['coord']['lon'],
                               'Max Temp':weather['main']['temp_max'],
                               'Humidity':weather['main']['humidity'],
                               'Cloudiness':weather['clouds']['all'],
                               'Wind Speed':weather['wind']['speed'],
                               'Current Description':weather['weather'][0]['description']
                               })

    except KeyError:
        print('No weather data, moving on...')
        pass
city_weather_ls
pd.DataFrame(city_weather_ls)
```

```python
WeatherPy_Data_df = pd.DataFrame(city_weather_ls)
WeatherPy_Data_df.to_csv('Weather Database/WeatherPy_Database.csv')
```
